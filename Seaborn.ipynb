## Data manipulation and visualization techniques 

## 1. Count plot

A count plot is is useful for categorical values. It is used to represent the frequency of different categories. The gender column  named **sex** contains categorical data in the Titanic data, i.e., men and women.

```python
# countplot() function has two attributes: data - The dataframe and x - The name of the column.

sns.countplot(x='sex',data=titanic)
```

We can see from the graph above that there were more men (about 580) than women (about 300) on the Titanic.

So the number of male passengers was significantly higher than the number of female passengers.







We can also divide the bars of the count plot according to another categorical variable. The color palette of the graph can also be customized. 

```python
# the attribute hue contains the name of the categorical column to split the bars which is survived here.
sns.countplot(x='sex', hue = 'survived', data = titanic,palette = 'Set1')
```

For survived column contains two values 0 for no and 1 for yes.

We observe from the graph that the number of survived male passengers was significantly less than the number of survived female passengers.

So, most of the dead was male passengers.

## 2. Kernel Density Estimate (KDE) Plot


A Kernel Density Estimate (KDE) Plot is used to plot the distribution of continuous data.

```python
sns.kdeplot(x = 'fare' , data = titanic , color = 'black')
```

The peak of the above graph is in between  5 and 10 so we can conclude that the fare for most passengers was between 5 and 10. 

## 3. Distribution plot

A Distribution plot is similar to a KDE plot. It is used to plot the distribution of continuous data.

displot() function combines the matplotlib hist function (with automatic calculation of a good default bin size) with the seaborn kdeplot() and rugplot() functions. 

```python
# kde: It is set to False by default. However, if you wish to plot a KDE graph on top of the bars, you can set it to True. 
# bins: the number of bins/bars. The lower the number, wider the bars and wider the intervals.

sns.displot(x = 'age', kde=True, bins = 5 , data =titanic)
```

The graph above tells us that most of the people on board the Titanic were in their early twenties.

```python
# You can also pass hue and palette as parameters to customize the graph.

sns.displot(x ='age', kde=True, bins = 5 , hue = titanic['survived'] , palette = 'Set2', data=titanic)
```

Most of the passengers were in their twenties, as were the survivors. 

## 4. Box Plot

A boxplot is a standard way of visualising the dataset based on the five-number summary: 
* the minimum ($Q_0$ or 0%), 
* the maximum ($Q_4$ or 100%), 
* the median of the sample ($Q_2$ or 50%), 
* the first quartile ($Q_1$ or 25%),
* third quartile ($Q_3$ or 75%). 




```python
# The below graph is a Boxplot
# and the different part of a boxplot
from IPython.display import Image
Image("/content/drive/MyDrive/boxplot.png")
```

In the titanic, there was three classes (first, second and third). 
Let us plot the boxplot of the different classes function of the age of passengers.

```python
# attribute showfliers is using to remove outliers
sns.boxplot(x='class',y='age',data=titanic, showfliers=False)
```

From the graph above, we can see that, on board the Titanic.

* In the first class, the median age of the passengers was about 38 years. Furthermore, 50% of the total passengers were between 28 and 50 years old, and the maximum age in the first class was 80 years.

* In the second class, the median age of the passengers was about 30 years. Furthermore, 50% of the total passengers were between 25 and 38 years old, the maximum age in the second class was around 55 years.

* In the third class, the median age of the passengers was about 25 years. Furthermore, 50% of the total passengers were between 19 and 33 years old, the maximum age in the second class was around 52 years.


In addition, the median age of first class passengers is higher than that of second class passengers, and the median age of second class passengers slightly higher than that of third class passengers.

## 5. Swarm plot

A swarm plot is a type of scatter plot used to represent the distribution of an attribute or the joint distribution of two attributes.

It allows for a better understanding of the distribution by showing all the data points and avoiding overlapping points. It also helps to understand how the data is distributed over a categorical attribute and how the continuous variable varies within a category.


In the following, we plot the categorical attribute **class** function of the continuous variable **age**.

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

We can observe from the above graph that there were more passengers in third class than in first and second class. In addition, there were more children under 15 years of age in the third class. 

From the graph above, we can add the attribute hue= survived to divide the passengers of the different classes into survivors and non-survivors in order to visualise in which class there were more survivors or deaths.

```python
sns.swarmplot(x='class',y='age',hue="survived", data=titanic,palette='Set2')
```

We can see from the graph above that most of the survivors of the Titanic were in first class and that almost all the passengers died in third class.




For the following plots, we will be working with the **iris dataset**. The iris dataset contains data related to flowerâ€™s petal size (petal length and petal width) and sepal size (sepal length and sepal width).

These features are used to classify the type of iris (Setosa, Versicolour, and Virginica). Below we will try to investigate the relation between the features.

## 6. Scatter plot

A scatter plot is used to observe relationships between variables. It uses dots to represent values for two different numeric variables. 

A scatter plot requires data for its x-axis and y-axis, and the position of each dot on the horizontal and vertical axis indicates values for an individual data point. 
 
 We can also pass a value for the hue parameter to color the dots based on a categorical column.





First, we will need to load the iris dataset.




```python
iris = sns.load_dataset('iris')
iris.head()
```

Let us plot the relationship between the variables sepal_length and petal_length divided into different types of species. 

```python
sns.scatterplot(x='sepal_length', y ='petal_length' ,
data = iris , hue = 'species')
```

In the graph above we can observe that:

* Iris flowers with sepal length < 6cm and petal length > 2cm is most likely setosa type; 

* Iris flowers with a petal length between 2cm and 5cm are most likely versicolor and may also be virginica; 

* Iris flowers with a petal length > 5cm are most probably virginica.


## 7. Joint plot

A Joint Plot is used to plot correlation between 2 variables (bivariate) and a univariate graph in the margins. 


```python
# Plot the relation between sepal_lenght and petal_lenght by default the plot is a scatter plot
# and the univariate graph is the histogram plot

sns.jointplot(x='sepal_length' , y ='petal_length', data = iris)
```

In the graph above, we can see that the sepal length of the different species increases with the petal length. Furthermore, using the histogram, we can see that there are more species with a petal length between 1 cm and 2 cm and those with a sepal length between 5.5 cm and 6.5 cm.

We can add more attributes, for better visualization.

```python
# kind define the type of plot to draw. 
# There are 6 kinds such as {â€œscatterâ€ | â€œkdeâ€ | â€œhistâ€ | â€œhexâ€ | â€œregâ€ | â€œresidâ€ }

sns.jointplot(x='sepal_length' , y ='petal_length',data = iris, kind="kde", hue="species")
```

## 8. Pair plots

A pair plot is a matrix of scatter plots that allows you to understand the pairwise relationship between different features in a dataset. It creates a grid of axes such that each features in the dataset will be split into a single row y-axis and a single column x-axis.

The way to interpret the matrix is as follows:

* The variable names are displayed on the outer edges of the matrix.

* The boxes along the diagonals display the histogram plot (by default) for each variable.

* All other boxes show a scatter plot (by default but it can be another plot) of the relationship between each pair of features.

```python
# the pair plot below using the  iris flower dataset
sns.pairplot(iris)
```

For instance, if you only want petal_width and petal_length, tha would create a 2x2 plot.

```python
# 
sns.pairplot(iris, vars=["petal_width", "petal_length"])
```

From the above grid, we can see that petal_width and petal_length are correlated.

In the following plot, we can change the shape of the distribution and display the data point divided by species

```python
#we can change the shape of the distribution of the diagonal using diag_kind attribute
sns.pairplot(iris, diag_kind="kde", hue="species")
```

## 9. Heat maps

A heat map is a data visualisation technique in which values are represented by colours in two dimensions. It allows the viewer to understand complex data sets.

Heat maps are adapted to numerical data but if the data set contain a categorical variable, we use a **correlogram** which is a variant of the heatmap that replaces each of the variables on the two axes with a list of numeric variables in the dataset.

The following code creates the correlation matrix between all the iris features. 

```python
# corr() is a correlation function
corr = iris.corr()
# correlation metric
corr
```

the below code Creates the heatmaps from correlation matrices.

```python
# Grid of colors 
# each color is a relationship between two variables

sns.heatmap(corr)
```

In this heat map, each cell represents the relationship between the intersecting variables, such as a linear correlation.

The values that are closest to the dark colour represent stronger positive  stronger positive correlations. Those that are closer to white brown represent stronger positive negative correlations. 

```python
# the attribute annot write the data value in each cell if true.
sns.heatmap(corr, annot=True)
```

* Petal length is highly correlated with petal width and sepal length; 

* sepal width is negatively correlated with the other three variables.
